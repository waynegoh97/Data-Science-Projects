```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
x=[2,4,6,8,10]
y=[3,6,9,12,15]
plt.plot(x,y)
plt.xlabel("Multiple of 2")
plt.ylabel("Multiple of 3")
plt.show()
```

```python
x = np.linspace(-1,1,50)
x
```

```python
y1 = x**4 + 1
y2 = 4**x + 1

plt.plot(x,y1)
plt.plot(x,y2,color='green',linewidth=1.5,linestyle='--')
plt.show()
```

```python
x=[11,22,33,44,55]
y=[9,18,27,36,45]

plt.scatter(x,y)
plt.xlabel("Multiples of 11")
plt.ylabel("Multiples of 9")
plt.show()
```

```python
x1=['apple','banana','chiku','dragon','kiwi']
y1=[23,45,33,12,52]

plt.bar(x1,y1)
plt.xlabel("Fruits")
plt.ylabel("Count")
plt.show()
```

```python
#Sample data
d = {'Name':['Ali','Chi','Da','Gau','Nick','Anu'],
     'Age':[20,22,25,32,19,21],
     'Gender':['M','F','F','M','M','F']}
df = pd.DataFrame(d)
df
```

```python
plt.boxplot(df['Age'])
plt.title("boxplot for Scores column")
```

```python
plt.pie(df['Age'],labels=df['Name'])
plt.show()
```

```python
c1 = np.random.normal(100,10,200)
c2 = np.random.normal(89,30,200)

data_list = [c1,c2]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.boxplot(data_list)
plt.show()
```

```python
#Create subplots
fig, ax = plt.subplots(2,2)
x = np.arange(1,10)
ax[0][0].plot(x,x*x)
ax[0][0].set_title("Square")
ax[0][1].plot(x,np.sqrt(x))
ax[0][1].set_title("Square Root")
ax[1][0].plot(x,np.exp(x))
ax[1][1].plot(x,np.log(x))
ax[0][0].grid(color='b',ls='-.',lw=0.25)
plt.show()
```
